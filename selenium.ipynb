```python
from selenium import webdriver
browser = webdriver.Edge()
```

```python
browser.get('https://www.amazon.in')
```

```python

input_text= browser.find_element(By.ID,'twotabsearchtextbox')
search_button = browser.find_element(By.ID,'nav-search-submit-button')
input_text.send_keys('Smartphone under 10000')
search_button.click()
sleep(3)
```

```python
mobiles = browser.find_elements(By.XPATH,".//div[@data-asin and @data-asin!=' ']")
```

```python
data =[]
for mobile in mobiles:
    try:
        description=mobile.find_element(By.CSS_SELECTOR,'a h2 span').text
    except:
        description='NaN'
    try:
        price = mobile.find_element(By.CLASS_NAME,'a-price-whole').text
    except:
        price='NaN'
    try:
        stars = mobile.find_element(By.CLASS_NAME,'a-icon-alt').get_attribute('innerHTML')
    except:
        stars='NaN'
    try:
        votes = mobile.find_element(
            By.XPATH,
            ".//a[@aria-label and contains(@aria-label,'ratings')]"
        ).get_attribute('aria-label')
    except:
        votes='NaN'
    try:
        link = mobile.find_element(
            By.XPATH,
            "../a[@class and contains(@class,'a-link-normal s-no-outline')]"
        ).get_attribute("href")

    except:
        link = 'NaN'
    print(description,'|' ,price,'|',stars,'|',votes,'|', link)
```

```python
df = pd.DataFrame(columns=['Price','Mobile Info','Ratings','Number of votes','Mobile Link'])
```

```python
df = df[df['Price']!= "NaN"]
df = df[df['Mobile Info']!= "NaN"]
df = df[df['Ratings']!= "NaN"]
df = df[df['Number of votes']!= "NaN"]
df = df[df['Mobile Link']!= "NaN"]
```
